<b>MATLAB-style Interface</b>

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.cos(x));


# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.sin(x))
```

<b>Object-oriented interface</b>

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(3)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

Scatter Plots with plt.plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(42)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(10), rng.rand(10), marker,
             label="marker='{0}'".format(marker))
    print(rng.rand(10))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

Scatter Plots with plt.scatter

```python
plt.scatter(x, y, marker='o');
```

The primary <b>difference</b> of plt.scatter from plt.plot is that it can be used to create scatter plots where the properties of <b>each individual point</b> (size, face color, edge color, etc.) <b>can be individually controlled</b> or mapped to data.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

<b>Гистограммы, разбиения по интервалам и плотноть</b>

```python
# X1 = np.random.randn(1000)
# X2 = np.random.randn(1000)
# X3 = np.random.randn(1000)
X1 = np.random.normal(1, 0.8, 1000)
X2 = np.random.normal(0.7, 0.9, 1000)
X3 = np.random.normal(0.4, 0.5, 1000)
```

```python
kwargs = dict(histtype='stepfilled',alpha=0.3, normed=True, bins=40)
plt.hist(X1, **kwargs, label='X1')
plt.hist(X2, **kwargs, label='X2')
plt.hist(X3, **kwargs, label='X3')
plt.legend()
```

<b>Выделим 5 интервалов и посчитаем кол-во точек в каждом из них</b>

```python
counts, bin_edges = np.histogram(X1, bins=5)
```

```python
counts
```

```python
bin_edges
```

```python
bin_edges[-6]
```

```python
counter = len(bin_edges)
for i in range(counter - 1):
    print('в интервале [',bin_edges[-counter],bin_edges[-counter+1],']',\
         counts[-counter+1], 'значений')
    counter -= 1
```

<b>Two-Dimensional Histograms and Binnings</b>

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

# Seaborn part

```python
import seaborn as sns
sns.set()
```

<b>Exploring Seaborn Plots</b>

matplotlib way

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], normed=True, alpha=0.5)
```

<b>kdeplot</b>

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

<b>Histograms and KDE</b> can be combined using distplot

```python
sns.distplot(data['x'])
sns.distplot(data['y']);
```

If we pass the full two-dimensional dataset to kdeplot, we will get a two-dimensional visualization of the data:

```python
sns.kdeplot(data);
```

<b>joint distribution</b>

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex')
```

<b>Pairplot</b>

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

<b>Фасетные гистограммы</b>

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

<b>Factor plots</b>

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill");
```

<b>joint plot - regression</b>

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```

<b>Столбчатые диаграммы</b>

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("year", data=planets, aspect=2,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=5)
```

order и hue!!!

```python
    g = sns.factorplot("year", data=planets, aspect=4.0, kind='count',
                       hue='method', order=range(2010, 2015))
    g.set_ylabels('Number of Planets Discovered')
```
